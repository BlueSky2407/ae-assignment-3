```python
# Read the Excel file
df = pd.read_excel("../data/Other Macroeconomic Indicators.xlsx", sheet_name="Weekly", skiprows=1)
print(df.columns)
# Convert 'Reporting Date' to datetime format
df[' Reporting Date'] = pd.to_datetime(df[' Reporting Date'], format='%d-%b-%Y')

# Sort in ascending order to ensure proper "last" value selection
df_sorted = df.sort_values(' Reporting Date', ascending=False)

# Create month-period column for grouping
df_sorted['Month'] = df_sorted[' Reporting Date'].dt.to_period('M')

# Get the last value for each month
monthly_last_values = df_sorted.groupby('Month').last()[[' Reporting Date', 'FOREIGN EXCHANGE RESERVES']].reset_index()
```

```python
monthly_last_values
```

```python
# Read the daily sheet from your Excel file
daily_df = pd.read_excel('../data/Other Macroeconomic Indicators.xlsx', sheet_name='Daily', skiprows=1)
print(daily_df.columns)
# Convert 'Reporting Date' to datetime
daily_df['Reporting Date'] = pd.to_datetime(daily_df['Reporting Date'], format='%d-%b-%Y')

# Sort by date ascending to ensure first observation is truly the first in the month
daily_df_sorted = daily_df.sort_values('Reporting Date', ascending=False)

# Create a 'Month' column as string for grouping
daily_df_sorted['Month'] = daily_df_sorted['Reporting Date'].dt.to_period('M').astype(str)

# Extract the first observation for each month for the last column
first_obs = (
    daily_df_sorted
    .groupby('Month')
    .first()
    .reset_index()[['Month', "RBI&apos;S REFERENCE RATE: INR PER USD"]]
)

(first_obs)
```

```python
first_obs.to_excel('../data/exchangerate.xlsx')
monthly_last_values.to_excel('../data/forex-reserves.xlsx')
raw_dataset_tb.to_excel('../data/tradebalance.xlsx')
```
